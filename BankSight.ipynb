```python
#1. Q1: How many customers exist per city, and what is their average account balance?
cursor.execute("USE banksight")
cursor.execute("""
select c.city, COUNT(c.Customer_id) AS customer_count, AVG(a.account_balance) AS Avg_Account_Balance
FROM Customers c
JOIN accounts a ON c.customer_id=a.customer_id
GROUP BY c.city
ORDER BY Avg_Account_Balance DESC;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
# Q2: Which account type (Savings, Current, Loan, etc.) holds the highest total
#balance?
cursor.execute("USE banksight")
cursor.execute("""
select c.account_type, SUM(a.account_balance) AS Total_account_balance
FROM Customers c
JOIN accounts a ON c.customer_id=a.customer_id
GROUP BY c.account_type
ORDER BY Total_account_balance DESC
LIMIT 1;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q3: Who are the top 10 customers by total account balance across all account
#types?
cursor.execute("USE banksight")
cursor.execute("""
select c.name, SUM(a.account_balance) AS Total_account_balance
FROM customers c
JOIN accounts a ON c.customer_id=a.customer_id
GROUP BY c.customer_id
ORDER BY Total_account_balance DESC
LIMIT 10;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q4: Which customers opened accounts in 2023 with a balance above ₹1,00,000?
#TRANSACTION BEHAVIOR
cursor.execute("USE banksight")
cursor.execute("""
select c.name, SUM(a.account_balance) AS Total_account_balance
FROM Customers c
JOIN accounts a ON c.customer_id=a.customer_id
WHERE YEAR(c.join_date)=2023
GROUP BY c.customer_id
HAVING SUM(a.account_balance) > 100000
ORDER BY c.name;

""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
# Q5: What is the total transaction volume (sum of amounts) by transaction type?
cursor.execute("USE banksight")
cursor.execute("""
select t.txn_type AS Transaction_Type, SUM(t.amount) AS Total_Transaction_Volume
FROM transactions t
GROUP BY Transaction_Type ;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q6: How many failed transactions occurred for each transaction type?
cursor.execute("USE banksight")
cursor.execute("""
select t.txn_type AS Transaction_Type, Count(t.status) 
FROM transactions t
WHERE t.status="Failed"
GROUP BY Transaction_Type ;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q7: What is the total number of transactions per transaction type?
cursor.execute("USE banksight")
cursor.execute("""
select t.txn_type AS Transaction_Type, COUNT(t.txn_id) AS Total_Transaction_Type 
FROM transactions t
GROUP BY t.txn_type ;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q8: Which accounts have 5 or more high-value transactions above ₹20,000?
#LOAN INSIGHTS
cursor.execute("USE banksight")
cursor.execute("""
select c.name,COUNT(t.txn_id) AS High_Value_Transaction_Count
FROM Customers c
JOIN transactions t ON c.customer_id=t.customer_id
WHERE t.amount > 20000
GROUP BY c.customer_id
HAVING COUNT(t.txn_id) >= 5
ORDER BY High_Value_Transaction_Count DESC;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q9: What is the average loan amount and interest rate by loan type (Personal,
#Auto, Home, etc.)?
cursor.execute("USE banksight")
cursor.execute("""
select l.Loan_Type,AVG(l.Loan_Amount),AVG(l.Interest_Rate)
FROM loans l
GROUP BY l.Loan_Type
ORDER BY l.Loan_Type;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q10: Which customers currently hold more than one active or approved loan?
cursor.execute("USE banksight")
cursor.execute("""
select l.Customer_ID, COUNT(l.Customer_ID) AS Active_Approved_Loan_Count
FROM loans l
WHERE l.Loan_Status IN ("Approved","Active")
GROUP BY l.Customer_ID
HAVING COUNT(l.Customer_ID) > 1
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q11: Who are the top 5 customers with the highest outstanding (non-closed)
#loan amounts?
cursor.execute("USE banksight")
cursor.execute("""
select l.Customer_ID, SUM(l.Loan_Amount) AS Active_Approved_Loan_Amount
FROM loans l
WHERE l.Loan_Status IN ("Approved","Active")
GROUP BY l.Customer_ID
ORDER BY Active_Approved_Loan_Amount DESC
LIMIT 5;
""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q12: What is the average loan amount per branch?
cursor.execute("USE banksight")
cursor.execute("""
select l.Branch, AVG(l.Loan_Amount) AS Average_Loan_Amount
FROM loans l
GROUP BY l.Branch
ORDER BY l.Branch;

""")
results=cursor.fetchall()
for row in results:
	print(row)
```

```python
#Q13: How many customers exist in each age group (e.g., 18–25, 26–35, etc.)?
cursor.execute("USE banksight")
cursor.execute(""" 
select 
    CASE 
        WHEN age BETWEEN 18 AND 25 THEN '18-25'
        WHEN age BETWEEN 26 AND 35 THEN '26-35'
        WHEN age BETWEEN 36 AND 45 THEN '36-45'
        WHEN age BETWEEN 46 AND 55 THEN '46-55'
        WHEN age BETWEEN 56 AND 65 THEN '56-65'
        ELSE '66+' 
    END AS Age_Group,
    COUNT(*) AS Customer_Count  
FROM customers
GROUP BY Age_Group
ORDER BY Age_Group;
""")
results=cursor.fetchall()
for row in results:
    print(row)
```

```python
#Q14: Which issue categories have the longest average resolution time?
cursor.execute("USE banksight")
cursor.execute("""
select i.issue_category, AVG(DATEDIFF(i.date_Opened,i.date_Closed)) AS Avg_Resolution_Time
FROM support_tickets i
GROUP BY i.issue_category
ORDER BY Avg_Resolution_Time DESC;
""")
results=cursor.fetchall()
for row in results:
    print(row)  
```

```python
#Q15: Which support agents have resolved the most critical tickets with high customer ratings (≥4)?
cursor.execute("USE banksight")
cursor.execute("""
select s.support_agent, COUNT(*) AS Num_Critical_Tickets_Resolved
FROM support_tickets s
WHERE s.priority = 'Critical' AND s.customer_rating >= 4
GROUP BY s.support_agent
ORDER BY Num_Critical_Tickets_Resolved DESC;
""")
results=cursor.fetchall()
for row in results:
    print(row)
```
